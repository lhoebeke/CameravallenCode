A function is created that lists all folders inside the drive folder, i.e. the individual deployments:

```python
gconnect.list_deployments()
```

As this is a generator, it is straightforward to loop through the deployments:

```python
for deployment in gconnect.list_deployments():
    print("Gdrive identifier: ", deployment['id'], ", agouti identifier: ", deployment['name'])
```

Just a single deployment looks as follows (dict):

```python
deployment
```

Hence, the `id` is a HASH created by google drive to identifiy each folder, whereas the `name` is the Agouti deployment identifier. To query files in gdrive, we need the gdrive `id`

For each of these deployments, we can enlists the images as well:

```python
for j, image in enumerate(gconnect.list_deployment_images(deployment['id'])):
    print("Gdrive identifier: ", image['id'], ", image name: ", image['name'])
    if j > 10: # Just breaking here to limit number of printed image information
        break
```
